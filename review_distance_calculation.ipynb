## Recognize internal vs. external

```python
def incoming_outgoing_timings(sent, heard, other_sent):
    l_out = []
    l_in = []
    for i, timing in enumerate(heard):
        has_incoming = min(np.abs(timing - np.array(other_sent))) < 2
        has_outgoing = min(np.abs(timing - np.array(sent))) < 2
        if has_incoming and not has_outgoing:
            l_in.append(timing)
        elif has_outgoing and not has_incoming:
            l_out.append(timing)
    return l_in, l_out
def closest(x, vec):
    res = vec[np.argmin(np.abs(np.array(vec) - x))]
    if abs(res - x) < 1:
        return res
    else:
        return None
```

```python
def distances(heard1, sent1, heard2, sent2):
    lin1, lout1 = incoming_outgoing_timings(sent1, heard1, sent2)
    lin2, lout2 = incoming_outgoing_timings(sent2, heard2, sent1)
    
    l = []
    for x in sorted(set(lin1)):
        m_in1 = x
        m_out2 = closest(m_in1, lout2)

        if not m_out2:
            continue
        l.append([])
        for y in sorted(lin2):
            n_in2 = y
            n_out1 = closest(n_in2, lout1)
            l[-1].append(-1)
            if not n_out1:
                continue

            distance = ((m_in1 - m_out2) + (n_in2 - n_out1)) * 343 / 2
            l[-1][-1] = distance
    return pd.DataFrame(l, columns=sorted(lin2 - lin2[0]))
```
